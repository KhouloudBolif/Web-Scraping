Save data On CSV file 

creation de file pour sauvgarder les donnes

```python
f = open("data/jobs.csv","w" ,encoding="utf-8")
header ="job_title, country ,Name of Company,type of job , Time posted \n"
f.write(header)
```

```python
for container in containers:
    job_title = container.find_all("h2", {"class": "css-m604qf"})
    title = job_title[0].text
    
    job_country = container.find_all("span", {"class": "css-5wys0k"})
    country = job_country[0].text.strip()
    
    job_companyName = container.find_all("a", {"class": "css-17s97q8"})
    cname = job_companyName[0].text.strip()
    
    job_Type = container.find_all("span", {"class": "css-1ve4b75 eoyjyou0"})
    type = job_Type[0].text.strip()
    
    job_TimePosted = container.find_all("div", {"class": "css-4c4ojb"})
    TimePosted = job_TimePosted[0].text.strip()

    # crire les informations dans le fichier
    f.write(title + ";" + country + ";" + cname + ";" + type + ";" + TimePosted + "\n")

# Fermer correctement le fichier
f.close()
```
